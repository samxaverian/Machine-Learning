```python
import numpy as np
import matplotlib.pyplot as plt
import random
import sklearn
import time
import math
```

```python
random.seed(5)
x = np.random.uniform(-1,1,(200,1))

x.shape
print(x[:10,:])
```

As we can see that is does not generate numbers which are uniformly spaced.
So we create our own X as below

```python
x.shape
```

```python
xx = []
for i in range(-100,100):
    xx.append(i*0.01)
X = np.array(xx)

X.shape
#print(X[:10])
```

```python
y = []

def createClass(x):
    for i in x:
        if((i>=-0.5 and i<0.1) or (i>0.5)):
            y.append(1)
        else:
            y.append(-1)
    return np.array(y)
```

```python
y = createClass(X)
plt.scatter(X,y)
```

```python
y.shape
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
degree = 9
y_pred3 = []
error_rate = []
X = pd.DataFrame(X)
#fig, axes = plt.subplots(3, 3, figsize=(15, 6), sharey=True)
for i in range(9):
    polyreg=make_pipeline(PolynomialFeatures(i),LinearRegression())
    polyreg.fit(X,y)
    y_pred = polyreg.predict(X)
    if(i==3):
        y_pred3 = y_pred
    error = (1/200)*np.sum(np.square(y-y_pred))
    error_rate.append(error)
    plt.plot(X,y_pred)
    plt.plot(X,np.sign(y_pred))
    plt.title("Polynomial regression with degree "+str(i))
    plt.show()
```

```python
plt.plot(error_rate)
plt.ylabel("Error")
plt.xlabel("degree")
plt.title("Error rate vs degree")
```
